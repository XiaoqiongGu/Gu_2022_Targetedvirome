```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "arial"
import matplotlib.ticker as ticker
from numpy import inf 
```

```python
input = '/Users/xiaoqiong/Dropbox/3_FlowSorting_Viralmetagenomics_Paper/2021-analysis-P1/dna_features_viewer/recruitmentplot'
```

```python
P1_MAG1 = pd.read_csv(input + '/P1/vs2-pass2-MAG-1.blst',sep='\t',header=None)
PST_MAG1 = pd.read_csv(input + '/PST/vs2-pass2-MAG-1.blst',sep='\t',header=None)
SST_MAG1 = pd.read_csv(input + '/SST/vs2-pass2-MAG-1.blst',sep='\t',header=None)
```

```python
print(PST_MAG1.shape)
print(SST_MAG1.shape)
print(P1_MAG1.shape)
```

```python
def seq_count(df):
    Final = np.zeros(46095)
    for i in range(df.shape[0]):
        start = df.iloc[i,8]
        end = df.iloc[i,9]
        MIN = min(start,end)
        MAX = max(start,end)
        A = np.repeat([0],MIN-1)
        B = np.repeat([1],MAX-MIN+1)
        C = np.repeat([0],46095-MAX)
        SUM = np.concatenate([A, B, C], axis=0)
        Final = Final + SUM
    return Final
```

```python
#with open('P1_seq.npy','wb') as f:
#    np.save(f,Final)
#Final = seq_count(P1_MAG1)
#Final.sum()/46095
p1_seq = np.load('P1_seq.npy')
print(p1_seq.sum()/46095)
pst_seq = seq_count(PST_MAG1)
print(pst_seq.sum()/46095)
sst_seq = seq_count(SST_MAG1)
print(sst_seq.sum()/46095)
```

```python
x=[*range(46095)]
```

```python
plt.figure(figsize=(5, 2.5), dpi=300)

plt.scatter(x,Final,s=0.5,label='P1')
plt.axhline(y=1136,linestyle='--',linewidth=0.8,color='#00008B')
plt.scatter(x,pst_seq,s=0.5,label='PST')
plt.axhline(y=1.26,linestyle='--',linewidth=0.8,color='#F04A00')
plt.scatter(x,sst_seq,s=0.5, label='SST')
plt.axhline(y=0.04,linestyle='--',linewidth=0.8,color='#006400')
plt.yscale('log')
plt.legend(fontsize=8,markerscale=2)
plt.xlabel('Position in genome (Mbp)')
plt.ylabel('Sequencing depth (X)')
plt.savefig(input+'/plots/combined.svg',dpi=300)
```

```python
def pseudo_log_scale(arr):
    arr = np.log10(arr)
    arr[arr == -inf] = 0
    return arr
```

```python
p1_seq_log = pseudo_log_scale(p1_seq)
pst_seq_log = pseudo_log_scale(pst_seq)
sst_seq_log = pseudo_log_scale(sst_seq)
```

```python
#plt.figure(, dpi=300)
a = p1_seq_log.sum()/46095
b = pst_seq_log.sum()/46095
c = sst_seq_log.sum()/46095

fig, ax = plt.subplots(figsize=(10, 5))
plt.scatter(x,p1_seq_log,s=0.5,label='P1')
plt.axhline(y=a,linestyle='--',linewidth=1,color='#00008B')
plt.scatter(x,pst_seq_log,s=0.5,label='PST')
plt.axhline(y=b,linestyle='--',linewidth=1,color='#F04A00')
plt.scatter(x,sst_seq_log,s=0.5, label='SST')
plt.axhline(y=c,linestyle='--',linewidth=1,color='#006400')
plt.legend(fontsize=12,markerscale=3,loc='upper left',frameon=False)
plt.xlabel('Position in genome (bp)',fontsize=12)
plt.ylabel('Sequencing depth (X)',fontsize=12)
plt.savefig(input+'/plots/combined_linearscale.pdf',dpi=100)
```
