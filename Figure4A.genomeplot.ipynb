```python
import pandas as pd
keys = pd.read_csv('data/Figure4A.annotations.tsv',sep='\t')
lst = keys.iloc[:49,:][['start_position','end_position','strandedness','viral_hit','virsorter_category']]
def func(x):
    if x == 1:
        return "6aa84f"
    elif x == 2:
        return "ffd700"
    elif x == 0:
        return "808080"
    
lst['color_code'] = lst['virsorter_category'].apply(func)
lstb = lst.T.to_dict()
lst.head()
```

```python
import os
#os.remove('P1_features.txt')
fi = open('data/P1_features.txt','a')
for x in lstb: 
    if str(lstb[x]['viral_hit']) =='nan':
        fi.write('GraphicFeature(start='+ str(lstb[x]['start_position']) + ', end=' + str(lstb[x]['end_position']) + 
         ', strand=' + str(lstb[x]['strandedness']) +
         ', color="#' + str(lstb[x]['color_code']) + '"),\n')
    else:
        fi.write('GraphicFeature(start='+ str(lstb[x]['start_position']) + ', end=' + str(lstb[x]['end_position']) + 
         ', strand=' + str(lstb[x]['strandedness']) +
         ', color="#' + str(lstb[x]['color_code']) + 
          '",label="' + str(lstb[x]['viral_hit'].split(' ',1)[1]) + 
           '"),\n')
fi.close()
```
